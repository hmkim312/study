## 07-2 열이름 관리하기
#### ebola 데이터 집합 살펴보기

```python
# ebola 데이터 불러오기
ebola = pd.read_csv('data/country_timeseries.csv')
print(ebola.columns)
ebola.iloc[:5, [0, 1, 2, 3, 10, 11]]
```

```python
# date와 day를 고정 후 피벗
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long
```

#### 열 이름 분리하고 데이터프레임에 추가하기

```python
# split 메서드를 사용하기
variable_split = ebola_long.variable.str.split('_')
variable_split
```

```python
# variable_split의 타입은?
type(variable_split)
```

```python
type(variable_split[0])
```

```python
variable_split
```

```python
#get 메서드를 사용하여 0,1번째 인덱스의 데이터를 추출
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)
status_values[:5]
```

```python
status_values[-5:]
```

```python
country_values[:5]
```

```python
country_values[-5:]
```

```python
# status, country 열이름으로 데이터프레임에 추가한 코드
ebola_long['status'] = status_values
ebola_long['country'] = country_values
ebola_long
```

```python
# concat 메서드로 데이터프레임에 열 추가
variable_split = ebola_long.variable.str.split('_', expand = True)
variable_split.columns = ['status', 'country']
ebola_parsed = pd.concat([ebola_long, variable_split], axis = 1)
ebola_parsed
```

## 07-3 여러 열을 하나로 정리하기

#### 기상 데이터의 여러 열을 하나로 정리하기 - melt, pivot_table 메서드

```python
# 기상데이터를 출력
weather = pd.read_csv('data/weather.csv')
weather.iloc[:5,:]
```

```python
# melt 메서드로 d1부터 피벗
weather_melt = pd.melt(weather, id_vars=[
                       'id', 'year', 'month', 'element'], var_name='day', value_name='temp')
weather_melt
```

```python
# pivot_table
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns = 'element',
    values = 'temp',
    dropna = False)
weather_tidy
```

```python
# index 재설정
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat
```

## 07-4 중복 데이터 처리하기

#### 빌보드 차트의 중복 데이터 처리하기

```python
# 빌보드 차트 불러오기
billboard = pd.read_csv('data/billboard.csv')
billboard_long = pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], 
                         var_name='week', value_name='rating')
print(billboard_long.shape)
billboard_long
```

```python
# Loser 노래 중복 확인
billboard_long[billboard_long.track == 'Loser']
```

```python
# 중복데이터를 가지고 있는 열을 새로운 데이터프레임에 저장
billboard_songs = billboard_long[['year','artist','track','time']]
billboard_songs.shape
```

```python
# 중복제거
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
# id 추가
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs
```

```python
# merge 메서드를 이용해 노래정보와 주간순위 데이터 합침
billboard_ratings = billboard_long.merge(billboard_songs, on = ['year','artist','track','time'])
print(billboard_ratings.shape)
billboard_ratings
```
