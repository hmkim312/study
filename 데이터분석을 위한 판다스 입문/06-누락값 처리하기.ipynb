#### 누락값 확인하기

```python
# numpy의 누락값 가져오기
import numpy as np
from numpy import NaN, NAN, nan
```

```python
# 누락값은 0과는 다른 개념
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
```

```python
# 누락값은 자신끼리 비교해도 False가 나옴 왜? 값 자체가 없어서
print(NaN == NaN)
```

```python
# isnull 메서드로 확인
import pandas as pd

print(pd.isnull(NaN))
print(pd.isnull(nan))
print(pd.isnull(NAN))
```

```python
# 누락값이 아닌것도 가능

print(pd.notnull(NaN))
print(pd.notnull(42))
print(pd.notnull('missing'))
```

#### 누락값이 생기는 이유알아보기

```python
# 누락값이 있는 데이터 집합을 연결할떄 누락값이 생기는 경우
visited = pd.read_csv('data/survey_visited.csv')
survey = pd.read_csv('data/survey_survey.csv')
visited
```

```python
survey
```

```python
# 데이터 연결 후 nan값 확인
vs = pd.merge(visited, survey, left_on='ident', right_on='taken')
vs
```

```python
# 데이터를 입력할때 누락값이 생기는 경우
num_legs = pd.Series({'goat' : 4, 'amoeda' : nan})
print(type(num_legs))
num_legs
```

```python
scientists = pd.DataFrame({
    'Name' : ['Rosaline Franklin', 'William Gosset'],
    'Occupation' : ['Chemist', 'Statistician'],
    'Born' : ['1920-07-25', '1876-06-13'],
    'Died' : ['1958-04-16', '1937-10-16'],
    'missing' : [NaN, nan]})
print(type(scientists))
scientists
```

```python
# 범위를 지정하여 데이터를 추출할떄 누락값이 생기는 경우
gapminer = pd.read_csv('data/gapminder.tsv', sep='\t')
gapminer
```

```python
life_exp = gapminer.groupby(['year'])['lifeExp'].mean()
life_exp
```

```python
# range 메서드를 이용해서 life_exp열에서 2000 ~2009데이터 추출, life_exp에는 없는 연도는 누락됨
# 현재는 loc에 nan값은 출력하지못하게 하게 업데이트하여 오류가 남
life_exp.loc[range(2000,2010),]
```

```python
# bool 타입으로 변경 후 출력
life_exp[life_exp.index > 2000]
```

#### 누락값의 개수 구하기

```python
# 샘플 데이터 확인
ebola = pd.read_csv('data/country_timeseries.csv')
ebola
```

```python
# count로 누락값이 아닌것 확인
ebola.count()
```

```python
# shpae[0]에 전체행의 데이터개수가 저장된점을 이용하여 nan값 구하기
num_rows = ebola.shape[0]
num_missing = num_rows - ebola.count()
num_missing
```

```python
# count_nonzero, isnull 메서드 조합으로 구하기
import numpy as np
np.count_nonzero(ebola.isnull())
```

```python
np.count_nonzero(ebola['Cases_Guinea'].isnull())
```

```python
# value_counts로도 가능
ebola.Cases_Guinea.value_counts(dropna = False).head()
```

#### 누락값 처리하기 - 변경, 삭제

```python
# 누락값 변경하기 0으로
ebola.fillna(0).iloc[0:10, 0:5]
```

```python
# fillna 메서드로 ffill을 주면 누락값이 나타기전의 값으로 변경됨
# 전의 행을 사용하여 누락값을 채웠으나, 0과 1은 앞이 NaN이라 변경이 안됨
ebola.fillna(method='ffill').iloc[0:10, 0:5]
```

```python
# method의 인자값을 bfill로 하면 누락값이 나타난 이후의 첫번째 값으로 누락값이 변경됨
# 위의 반대이나, 마찬가지로 맨 마지막에 있는nan값은 처리하지못함
ebola.fillna(method='bfill').iloc[0:10, 0:5]
```

```python
# interplate는 누락값 양쪽에 있는 값을 이용하여 중간값을 구한다음 누락값을 처리
ebola.interpolate().iloc[0:10,0:5]
```

```python
# 누락값 삭제, dropna
print('row데이터', ebola.shape)
ebola_dropna = ebola.dropna()
print('nan값 삭제 후 데이터', ebola_dropna.shape)
```

#### 누락값이 포함된 데이터 계산

```python
# 누락값이 포함된 열을 가지고 합을 계산하면?
ebola['Cases_multiple'] = ebola['Cases_Guinea'] + ebola['Cases_Liberia'] + ebola['Cases_SierraLeone']
ebola['Cases_multiple']
```

```python
# 값 확인
ebola_subset = ebola.loc[:,['Cases_Guinea','Cases_Liberia','Cases_SierraLeone','Cases_multiple']]
ebola_subset
```

```python
# sum 메서드에 skipna = True를 주어서 nan값을 무시
print(ebola.Cases_Guinea.sum(skipna= True))
print(ebola.Cases_Guinea.sum(skipna= False))
```
