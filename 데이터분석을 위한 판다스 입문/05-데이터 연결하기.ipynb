## 05-2 데이터 연결 기초

#### 데이터 연결하기

```python
# concat 메서드로 데이터 연결
import pandas as pd
df1 = pd.read_csv('data/concat_1.csv')
df2 = pd.read_csv('data/concat_2.csv')
df3 = pd.read_csv('data/concat_3.csv')
row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
# 행 데이터 추출
row_concat.iloc[3,]
```

```python
# 데이터 프레임에 시리즈 연결하기 (열로 추가됨)
new_row_series = pd.Series(['n1','n2','n3','n4'])
pd.concat([df1, new_row_series])
```

#### 행 1개로 구성된 데이터프레임 생성하여 연결하기

```python
# 1개의 행을 가지는 데이터 프레임을 생성하여 df1을 연결
new_row_df = pd.DataFrame([['n1','n2','n3','n4']], columns=['A','B','C','D'])
new_row_df
```

```python
pd.concat([df1, new_row_df])
```

```python
# append로도 가능함
df1.append(new_row_df)
```

```python
# append와 dict를 사용하여 행을 연결
data_dict = {'A': 'n1', 'B': 'n2', 'C': 'n3', 'D': 'n4'}
df1.append(data_dict, ignore_index=True)
```

#### 다양한 방법으로 데이터 연결하기

```python
# ignore_index 인자 사용하기(인덱스를 0부더 다시 지정)
row_concat_i = pd.concat([df1, df2, df3], ignore_index= True)
row_concat_i
```

```python
# 열 방향으로 데이터 연결하기,  axis = 1
col_index = pd.concat([df1, df2, df3], axis=1)
col_index
```

```python
# 같은 이름의 열은 모두 가져옴
col_index['A']
```

```python
# 새로운 열 추가
col_index['new_col_list'] = ['n1', 'n2', 'n3', 'n4']
col_index
```

```python
# 열 이름 초기화도 ignore_index
pd.concat([df1, df2, df3], axis=1, ignore_index=True)
```

```python
# 공통열과 공통 인덱스만 연결
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['E', 'F', 'G', 'H']
df3.columns = ['A', 'C', 'F', 'H']
print(type(df1))
df1
```

```python
print(type(df2))
df2
```

```python
print(type(df3))
df3
```

```python
# concat으로 연결 - 열이 없는것은  Nan값으로 연결됨
pd.concat([df1, df2, df3])
```

```python
# 공통열만 연결 join = 'inner'
# df1, df2, df3의 모두가진 공통 열이 없음
pd.concat([df1,df2,df3], join='inner')
```

```python
# df1과 df3은 A,C가 공통열
pd.concat([df1,df3], join='inner')
```

```python
#df2, df3은 F,H가 공통열
pd.concat([df2,df3], join='inner')
```

```python
# 행방향으로 연결
df1.index = [0,1,2,3]
df2.index = [4,5,6,7]
df3.index = [0,2,5,7]
df1
```

```python
df2
```

```python
df3
```

```python
# 행방양으로 하면 nan값이 나옴 - df1,df2,df3이 모두 공통으로 가지는 열이 없음
col_concat = pd.concat([df1,df2,df3], axis=1)
col_concat
```

```python
# df1,df3이 가지는 공통된 행만 연결
pd.concat([df1, df3], axis=1, join='inner')
```

```python
# df2,df3이 가지는 공통된 행만 연결
pd.concat([df2, df3], axis=1, join='inner')
```

## 05-3 데이터 연결 마무리

#### merge 메서드 사용하기

```python
person = pd.read_csv('data/survey_person.csv')
site = pd.read_csv('data/survey_site.csv')
survey = pd.read_csv('data/survey_survey.csv')
visited = pd.read_csv('data/survey_visited.csv')
person
```

```python
site
```

```python
survey
```

```python
visited
```

```python
# visited의 데이터프레임의 일부만 뗴어 실습에 사용
visited_subset = visited.loc[[0,2,6],]
```

```python
# merge
o2o_merge = site.merge(visited_subset, left_on= 'name', right_on='site')
o2o_merge
```

```python
# site, visited 데이터프레임의 데이터 연결
m2o_merge = site.merge(visited, left_on= 'name', right_on='site')
m2o_merge
```

```python
# person과 survey, visited와 surver를 merge
ps = person.merge(survey, left_on='ident', right_on='person')
vs = visited.merge(survey, left_on='ident', right_on='taken')
ps
```

```python
vs
```

```python
# left_on과 right_on의 갯수는 여러개여도 됨
ps_vs = ps.merge(vs, left_on=['ident', 'taken', 'quant', 'reading'], right_on=[
                 'person', 'ident', 'quant', 'reading'])
ps_vs
```

```python
# ps_vs 데이터프레임의 첫번째 행을 보면 중복된 행엔 _x, _y가 붙어있음
ps_vs.loc[0, ]
```
