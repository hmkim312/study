## 165 스택 구현하기
- 파이썬 리스트의 특성을 이용해 스택을 구현하는 방법을 학습
- stack : 나중에 저장된 자료가 먼저 추출될 수 있도록 되어 있는 나열 구조

```python
mystack = []
def putdata(data):
    global mystack
    mystack.append(data)
    
def popdata():
    global mystack
    if len(mystack) == 0:
        return None
    return mystack.pop()

putdata('데이터1')
putdata([3,4,5,6])
putdata(12345)

print(f'<스택상태> : {mystack}')

ret = popdata()
while ret != None:
    print(f'스택에서 데이터 추출 : {ret}')
    print(f'<스택상태> : {mystack}')
    ret = popdata()
```

## 166 문장에 나타나는 문자 빈도수 계산하기
- 문장에 등장하는 문자의 빈도수를 계산하는 방법 학습

```python
def getTextFreq(fliename):
    with open(fliename, 'r') as f:
        text = f.read()
        fa = {}
        for c in text:
            if c in fa:
                fa[c] += 1
            else:
                fa[c] = 1
    return fa


ret = getTextFreq('sample_data/mydata.txt')
ret = sorted(ret.items(), key=lambda x: x[1], reverse=True)
for c, freq in ret:
    if c == '\n':
        continue
    print(f'[{c}] -> {freq}회 나타남')
```

## 168 파일에서 특정 단어 개수 계산하기
- 텍스트 파일에서 특정 단어의 개수를 구하는 방법을 학습

```python
def countWord(filename, word):
    with open(filename, 'r') as f:
        text = f.read()
        text = text.lower()
        pos = text.find(word)
        count = 0
        while pos != -1:
            count += 1
            pos = text.find(word, pos + 1)
    return count

word = input('mydata.txt에서 개수를 구할 단어를 입력하세요')
word = word.lower()
ret = countWord('sample_data/mydata.txt', word)
print(f'[{word}]의 개수 : {ret}')
```
