## 170 URL에 접속하여 HTML 페이지 화면에 출력하기
- URL을 이용해 인터넷에 접속하여 해당 웹페이지의 HTML을 화면에 출력하는 방법을 학습

```python
from urllib.request import urlopen

url = 'https://www.python.org/'
with urlopen(url) as f:
    doc = f.read().decode()
    print(doc)
```

## 171 URL에 접속하여 HTML 페이지를 파일로 저장하기
- URL을 이용해 인터넷에 접속하여 해당 웹페이지의 HTML을 파일로 저장하는 방법을 학습

## 172 인터넷에 있는 이미지를 내 PC로 저장하기
- 이미지 URL을 이용해 인터넷에 있는 이미지를 내 PC로 저장하는 방법을 학습

```python
from urllib.request import urlopen

imgurl = '적절한 이미지 url'
imgname = imgurl.split('/')[-1]

try :
    with urlopen(imgurl) as f:
        with open(imgname, 'wb') as h:
            img = f.read()
            h.write(img)
except Exception as e:
    print(e)
```

## 173 인터넷에 있는 대용량 파일을 내pc로 저장하기
- URL을 이용해 인터넷에 있는 용량이 큰 파일을 내PC로 저장하는 방법을 학습ㅡ

```python
from urllib.request import urlopen

BUFSIZE = 256 * 1024

fileurl = 'https://www.python.org/ftp/python/3.7.2/python-3.7.2.exe'
filename = 'sample_data/' + fileurl.split('/')[-1]

try:
    with urlopen(fileurl) as f:
        with open(filename, 'wb') as h:
            buf = f.read(BUFSIZE)
            while buf:
                h.write(buf)
                buf = f.read(BUFSIZE)
except Exception as e:
    print(e)
```

## 174 10MB 파일을 1MB 파일 10개로 분리하기
- 용량이 큰 파일을 용량이 작은 여러개의 파일로 분리하는 방법을 학습

```python
filename = 'sample_data/python-3.7.2.exe'
subsize = 1024*1024*3 # 3mb
suffix = 0

with open(filename, 'rb') as f:
    buf = f.read(subsize)
    while buf:
        subfilename = filename + '_' + str(suffix)
        with open(subfilename, 'wb') as h:
            h.write(buf)
            print(f'{subfilename} 완료')
            
        buf = f.read(subsize)
        suffix += 1
```

## 175 1mb 파일 10개를 합쳐서 10mb파일로 만들기
- 분리한 여러개의 파일을 합쳐 원본 파일로 만드는 방법에 대해 학습

```python
BUFSIZE = 256 * 1024
merge_filename = 'sample_data/ret.exe'
filelist = ['sample_data/python-3.7.2.exe_' + str(x) for x in range(9)]

with open(merge_filename, 'wb') as f:
    for filename in filelist:
        print(f'{filename}합치는 중')
        with open(filename, 'rb') as h:
            buf = h.read(BUFSIZE)
            while buf:
                f.write(buf)
                buf = h.read(BUFSIZE)
                
print('파일합치기가 완료되었습니다.')
```

## 176 파일을 zip 압축파일로 만들기
- 하나의 파일을 zip으로 압축하는 방법에 대해 학습

```python
from zipfile import *

def compresszip(zipname, filename):
    print(f'{filename} -> {zipname} 압축중')
    with ZipFile(zipname, 'w') as ziph:
        ziph.write(filename)
        
    print('압축 완료')

filename = 'sample_data/mydata.txt'
zipname = filename +'.zip'
compresszip(zipname, filename)
```

## 177 디렉터리를 하나의 zip압축 파일로 만들기
- 디렉터리에 있는 모든 하위 디렉터리와 파일들을 하나의 zip 압축파일로 만드는 방법에 대해 학습
- os.walk : 주어진 디렉터리를 탐색하는 함수

```python
from zipfile import *
import os


def compressall(zipname, folder):
    print(f'{folder} -> {zipname} 압축중')
    with ZipFile(zipname, 'w') as ziph:
        for dirname, subdir, files in os.walk(folder):
            for file in files:
                ziph.write(os.path.join(dirname, file))


folder = 'sample_data'
zipname = folder + '.zip'
compressall(zipname, folder)
```

## 178 zip파일 압축풀기
- zip 파일을 푸는 방법을 학습

```python
from zipfile import *

def extractzip(zipname):
    with ZipFile(zipname, 'r') as ziph:
        ziph.extractall()
        print(f'{zipname}가 성공적으로 추출되었습니다.')
        
extractzip('sample_data.zip')
```

## 179 로또번호 추출기 만들기
- 1 ~ 45번 숫자에서 6개의 임의의 숫자로 이루어진 로또 번호 추출기를 생성
- pop()은 리스트의 맨 마지막 요소를 돌려주고 그 요소는 삭제한다.

```python
from random import shuffle
from time import sleep

gamenum = input('게임 회수를 입력하세요')

for i in range(int(gamenum)):
    balls = [x + 1 for x in range(45)]
    ret = []
    
    for j in range(6):
        shuffle(balls)
        number = balls.pop()
        ret.append(number)
    ret.sort()
    print(f'{i+1}회 로또번호',  end = '')
    print(ret)
    sleep(1)
```

## 180 남녀 파트러 정해주기 프로그램 만들기
- 같은 수의 남녀 모임에서 파트너를 정해주는 프로그램을 만듬

```python
from random import shuffle

male = ['슈퍼맨', '심봉사', '로미오', '이몽룡', '마루치']
female = ['원더우먼', '뺑덕', '줄리엣', '성춘향', '아라치']
shuffle(male)
shuffle(female)
couples = zip(male, female)

for i, couple in enumerate(couples):
    print('커플%d: [%s]-[%s]' % (i+1, couple[0], couple[1]))
```
