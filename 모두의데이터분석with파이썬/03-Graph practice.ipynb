# 제목, 범례 만들기

```python
plt.title("Age of Body weight") # 제목 정하기
plt.plot([20,23,25,27,29,31,33],[52,55,61,65,72,79,65],label = "Body weight") # label (선의 이름) 정하기
plt.legend() # lable 값을 범례로 바꾸기
#pitl. legend(loc = 숫자) 숫자를 넣어 주면 범례의 위치가 바뀜
plt.show()
```

# 그래프 색상 바꾸기

```python
plt.title("color")
plt.plot([10,20,30,40], color = "skyblue", label = "skyblue")
plt.plot([40,30,20,10], "pink" , label = "pink")
plt.legend(loc = 5)
plt.show()
```

# 그래프 선 모양 바꾸기

```python
plt.title("Linestyle")
plt.plot([10,20,30,40], color = "red", linestyle = "--", label = "Dashed") # 빨간색 Dashed 그래프
plt.plot([40,30,20,10], color = "green", linestyle = ":" ,  label = "Dotted") # 초록색 Dotted 그래프
plt.legend(loc = 5)
plt.show()
```

# 마커 모양 바꾸기

```python
plt.title("Marker")
plt.plot(a, "b.", label = "circle") #"b."은 블루 서클(.)을 의미 해당 함수를 쓸때 color =  을 쓰면 오류남
plt.plot(b, "r^", label= "triangle up")  # "r ^"은 레드의 삼각형(^)을 의미
# 색상, 선 모양, 마커 모양 순으로 작성하면 됨. plt.plot([10,20,30,40], "r.--")
plt.legend( loc = 5)
plt.show()
```
