## 연산자 사용하기

```python
# 산술연산자
print(3 * 10)  # 곱하기
print(3 ** 10)  # 제곱
print(3 % 10)  # 나머지
print(3 / 2)  # 나눗셈
print(3 // 2)  # 몫
```

```python
# 비교 연산자
print(10 >= 3) # 크거나 같다
print(10 <= 3) # 작거나 같다
print(10 == 3) # 같다
print(10 != 3) # 같지 않다.
print(3 % 2 == 1) # 같다
```

```python
# 논리 연산자
print(1 == 1 and 2 != 1) # True and True
print(10 % 2 != 0 and 1 + 1 > 0) # True and False
```

```python
# 논리 연산자
print(not 10 > 5) # not True
print(not 10 ==5) # not False
print(not 0) # 0은 False
print(not 10) # 0을 제외한 숫자도 True
```

## 반복과 선택

```python
# for 반복문
names = '파이쏭'
for name in names:
    print(name)
```

```python
# for 반복문(리스트)
names = ['쵸파', '상디', '루피', '조로']
for name in names:
    print(name)
```

```python
# for 반복문
for i in [0, 1, 2, 3]:
    print(i ** 2)
```

```python
# for 반복문(range)
for i in range(100):
    print(i ** 2)
```

```python
# if 조건문
if 10 > 0:
    print('정답입니다')
```

```python
# if 조건문
if 10 != 0 and 5 % 2 == 1:
    print('정답입니다')
```

```python
# if 조건문
passwd = int(input('비밀번호 4자리를 입력하세요 : '))
if passwd == 1531:
    print('비밀번호가 일치합니다')
```

```python
# if 조건문
for i in range(10000):
    if i == 1531:
        print(f'{i}은 비밀번호가 일치 합니다.')
```

```python
# if 조건문
passwd = int(input('비밀번호 4자리를 입력하세요 : '))
if passwd == 1531:
    print('비밀번호가 일치합니다')
else :
    print('비밀번호가 틀렸습니다.')
```

## 순서 있는 저장 공간 리스트

```python
for i in [0, 1, 2, 3]:
    print(i ** 2)
```

```python
names = ['쵸파', '상디', '루피', '조로']
for name in names:
    print(name)
```

```python
# 리스트에 저장된 위치로 값에 접근하기(indexing)
names = ['쵸파', '상디', '루피', '조로']
print(names[1])
```

```python
# 리스트에 저장된 위치로 값에 접근하기(indexing), 0부터 시작
names = ['쵸파', '상디', '루피', '조로']
print(names[0])
print(names[1])
print(names[2])
print(names[3])
```

```python
# 리스트에 저장된 위치로 값에 접근하기(indexing)
names = ['쵸파', '상디', '루피', '조로']
print(names[-1]) # 맨뒤에 것
```

```python
# 리스트에 저장된 위치로 데이터의 일부 자르기 (slicing)
names = ['쵸파', '상디', '루피', '조로']
print(names[0:2])
print(names[1:3])
print(names[1:])
print(names[:])
```

```python
# 리스트에 값 추가하기
names = ['쵸파', '상디', '루피', '조로']
names.append('나미')
print(names)
```

```python
# 리스트의 길이 출력 하기
print(len(names))
print(len('data analysis for everyone'))
```

```python
# 종합
names = ['쵸파', '상디', '루피', '조로']
names.append('해적왕')
for name in names:
    if len(name) > 2:
        print(name, '입니다.')
```
