```python
import csv
f = open("..\data\SeoulTemperature_1904~2019.csv")
data = csv.reader(f)
next(data)
result = [] # 최고기온을 저장할 리스트

for row in data :
    if row[-1] != "" :# 최고 기온 데이터 값이 존재한다면 
        result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가
print(result)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,2))  # Graph 사이즈
plt.plot(result, 'r') 
plt.show()
```

# 내 생일의 기온 데이터를 보자

```python
import csv
import matplotlib.pyplot as plt

f = open("data\SeoulTemperature_1904~2019.csv")
data = csv.reader(f)
header = next(data)
high = [] # 최고 기온 값을 high 리스트에 저장
low = []  # 최저 기온 값을 low 리스트에 저장

for row in data :
         if row[-1] !="" and row[-2] !="" : # 최고 기온 값과 최저 기온값이 존재한다면
             if 1983 <= int(row[0].split("-")[0]) : # 1983년 이후 데이터 라면
                if row[0].split("-")[1] == "03" and row[0].split("-")[2] == "12" : # 3월 12일 이라면
                    high.append(float(row[-1])) # 최고 기온 값을 higt 리스트에 저장
                    low.append(float(row[-2])) # 최저 기온값을 low 리스트에 저장
plt.plot(high, "hotpink")
plt.plot(low, "skyblue")
plt.show()
```

# 제목, 범례, 사이즈 추가 및 수정

```python
import csv
import matplotlib.pyplot as plt

f = open("data\SeoulTemperature_1904~2019.csv")
data = csv.reader(f)
header = next(data)
high = [] # 최고 기온 값을 high 리스트에 저장
low = []  # 최저 기온 값을 low 리스트에 저장
m, d = input("생일을 입력하세요 Ex.03-12").split("-")
print(m,"월",d,"일의", "1983년 부터 최고기온과 최저 기온은 아래와 같습니다")

for row in data :
         if row[-1] !="" and row[-2] !="" : # 최고 기온 값과 최저 기온값이 존재한다면
             if 1983 <= int(row[0].split("-")[0]) : # 1983년 이후 데이터 라면
                if row[0].split("-")[1] == m and row[0].split("-")[2] == d : # 3월 12일 이라면
                    high.append(float(row[-1])) # 최고 기온 값을 higt 리스트에 저장
                    low.append(float(row[-2])) # 최저 기온값을 low 리스트에 저장
                    
plt.figure(figsize = (15, 5)) # 그래프 사이즈 변경
plt.rc('font', family="Malgun Gothic") # 한글 제목을 위한 폰트 설정
plt.rcParams["axes.unicode_minus"] = False # 마이너스 기호 깨짐 방지
plt.title("내 생일의 기온 변화 그래프")# 제목 설정
plt.plot(high, "hotpink", label = "Higt")
plt.plot(low, "skyblue", label = "Low")
plt.legend()
plt.show()
```
