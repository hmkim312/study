## pandas로 인구 구조 분석하기
- 데이터를 읽어오기
    - 전체 데이터를 총 인구수로 나누어 비율로 변환
    - 총 인구수와 연령구간 인구수를 삭제
- 궁금한 지역의 이름을 입력 받는다.
- 궁금한 지역의 인구 구조를 저장한다.
- 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역을 찾는다.
    - 전국의 모든 지역 중 한곳(B)를 선택 한다.
    - 궁금한 지역 A의 0세 인구 비율에서 B의 0세 인구 비율을 뺀다.
    - 위를 100세 이상 인구수에 해당하는 값까지 반복한 후 차이의 제곱을 모두 더한다.
    - 전국의 모든 지역에 대해 반복하며 그 차이가 가장 작은 지역을 찾는다.
- 가장 비슷한 곳의 인구 구조와 궁금한 지역의 인구 구조를 시각화 한다.

#### 데이터 읽어오기

```python
import pandas as pd
df = pd.read_csv('data/age_2019.02.csv', encoding='cp949', index_col=0)
df.head()
```

```python
# 연령별 인구 숫자를 비율로 바꾸고, 총인구수와 연령구간 인구수 데이터를 삭제
import pandas as pd
df = pd.read_csv('data/age_2019.02.csv', encoding='cp949', index_col=0)
df = df.div(df['2019년02월_계_총인구수'], axis=0)  # 전체 데이터를 총 인구수로 나눠서 비율로 변환
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']  # 총인구수, 연령구인구수 열 삭제
col = [i[11:] for i in df.columns]
df.columns = col
df
```

```python
# 궁금한 지역 이름 입력받고 해당 지역의 인구 구조 저장하기
name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name) # 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2 = df[a]
df2
```

```python
# 그래프로 그리기
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font', family = 'Malgun Gothic')
df2.T.plot()
plt.show()
```

#### 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 시각화하기

```python
import numpy as np

# 궁금한 지역 A의 인구 비율에서 B의 인구 비율을 뺀다.
x = df.sub(df2.iloc[0], axis=1)

# A의 인구 비율에서 B의 인구 비율을 뺀 값의 제곱 값을 모두 더한다.
y = np.power(x, 2)
z = y.sum(axis=1)
```

```python
# 상위 5곳지역을 보기
i = z.sort_values().index[:5]
```

```python
# 꺽은선 그래프로 그려주기
df.loc[i].T.plot()
plt.show()
```

```python
# 종합
import numpy as np

# 궁금한 지역 A의 인구 비율에서 B의 인구 비율을 뺀다.
x = df.sub(df2.iloc[0], axis=1)

# A의 인구 비율에서 B의 인구 비율을 뺀 값의 제곱 값을 모두 더한다.
y = np.power(x, 2)
z = y.sum(axis=1)

# 상위 5곳지역을 보기
i = z.sort_values().index[:5]

# 꺽은선 그래프로 그려주기
df.loc[i].T.plot()
plt.show()
```

```python
# 한줄로 줄이기
import matplotlib.pyplot as plt
%matplotlib inline

df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
```

#### 우리 동네와 인구 구조가 비슷한 지역들을 그래프로 나타내기(pandas 사용)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family = 'Malgun Gothic')
df = pd.read_csv('data/age_2019.02.csv', encoding='cp949', index_col=0)
df = df.div(df['2019년02월_계_총인구수'], axis=0)  # 전체 데이터를 총 인구수로 나눠서 비율로 변환
del df['2019년02월_계_총인구수'], df['2019년02월_계_연령구간인구수']  # 총인구수, 연령구인구수 열 삭제
col = [i[11:] for i in df.columns]
df.columns = col

name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)
df2 = df[a]

df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()

plt.show()
```
