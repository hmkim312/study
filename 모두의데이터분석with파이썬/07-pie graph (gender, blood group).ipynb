```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# 원 그래프 그리기 - pie 함수

```python
plt.pie([10, 20, 6, 10])
plt.show()
```

## 동그란 원으로 그리기

```python
size = [2441, 2312, 1031, 1233]
plt.axis('equal')
plt.pie(size)
plt.show()
```

## 레이블, 비율 범례 추가하기

```python
plt.rc("font", family = "Malgun Gothic") # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233] # 데이터
label = ["A형", "B형", "O형", "AB형"] # 레이블
plt.axis("equal")
plt.pie(size, labels = label, autopct = "%.1f%%") # autopct = auto percent, %를 보여줌
plt.legend() #범례추가
plt.show()
```

# 색 및 돌출 효과 정하기

```python
plt.rc('font', family='Malgun Gothic')
size = [2441, 2312, 1031, 1233]
label = ['A형', 'B형', 'O형', 'AB형']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink'] # 색상 리스트
plt.axis('equal')
plt.pie(size, labels = label, autopct = '%.1f%%', colors = color, explode = (0, 0, 0.1, 0)) # explode 로 돌출 설정
plt.legend()
plt.show()
```

# 신도림의 성별 인구 비율 표현하기

```python
import csv

f = open('../data/age_gender2_2019.02.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요')

for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :
            m += int(row[i+3]) # 3 ~ 103 번 데이터 더하기
            f += int(row[i+106]) # 106 ~206번 데이터 더하기
        break
size.append(m) # 남성 합계 데이터 리스트에 추가
size.append(f) # 여성 합계 데이터 리스트에 추가
print(size)
```

```python
import csv

f = open('../data/age_gender2_2019.02.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요')

for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :
            m += int(row[i+3]) # 3 ~ 103 번 데이터 더하기
            f += int(row[i+106]) # 106 ~206번 데이터 더하기
        break
size.append(m) # 남성 합계 데이터 리스트에 추가
size.append(f) # 여성 합계 데이터 리스트에 추가
print(size)

plt.rc('font', family = 'Malgun Gothic')
color = ['crimson', 'darkcyan'] # 색상 설정
plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct = '%.1f%%', colors = color, startangle = 90)
plt.title(name + " 지역의 남녀 성별 비율") # 제목 설정
plt.show()
```
