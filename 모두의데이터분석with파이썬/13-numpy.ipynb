## matplot 홈페이지의 numpy

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

t = []
p2 = []
p3 = []
for i in range(0, 50, 2):
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

## numpy 라이브러리를 활용해 그래프 그리기

```python
# numpy를 활용한 코드
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

dice = np.random.choice(6, 10)

plt.hist(dice, bins = 6)
plt.show()
```

```python
# unit6으로 만든 코드
import matplotlib.pyplot as plt
import random
%matplotlib inline

dice = []
for i in range(10):
    dice.append(random.randint(1,6))
plt.hist(dice, bins = 6)
plt.show()
```

```python
# numpy를 활용한 코드 (2) - 1000000번 해보기 + 확률
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

dice = np.random.choice(6, 1000000, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])

plt.hist(dice, bins=6)
plt.show()
```

```python
# numpy를 사용한 버블 차트 그리기

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.random.randint(10, 100, 200)
y = np.random.randint(10, 100, 200)
size = np.random.rand(100) * 100

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# unit 10에서 사용한 버블 차트 그리기 코드

import matplotlib.pyplot as plt
import random
%matplotlib inline

x = []
y = []
size = []

for i in range(200):
    x.append(random.randint(10, 100))
    y.append(random.randint(10, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

#### numpy 라이브러리를 사용하여 재미있는 버블차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
size = np.random.rand(100) * 100

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 + mask2]
y = y[mask1 + mask2]

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
