#### 출근 시간대 사람들이 가장 많이 타고 내리는 역은 어디일까?

```python
# 3시간동안 가장 많이 들어오는 역 확인

import csv
f = open('data/subwaytime_2020.05.csv')
data = csv.reader(f)
next(data)
next(data)

mx = 0  # 최대값을 저장할 변수
mx_station = ''  # 최대값을 가지는 역 이름 저장 변수
for row in data:
    row[4:] = map(int, row[4:])
    a = row[11:16:2] # 하차시간
    if sum(a) > mx:
        mx = sum(a)
        mx_station = row[3] + '('+row[1]+')'
print(mx_station, mx)
```

#### 밤 11시에 사람들이 가장 많이 타는 역은 어디일까?

```python
import csv
# 3시간동안 가장 많이 들어오는 역 확인

import csv
f = open('data/subwaytime_2020.05.csv')
data = csv.reader(f)
next(data)
next(data)
t = int(input('몇 시의 승차 인원이 가장 많은 역이 궁금하신가요'))

mx = 0  # 최대값을 저장할 변수
mx_station = ''  # 최대값을 가지는 역 이름 저장 변수
for row in data:
    row[4:] = map(int, row[4:])
    a = row[4+(t-4)*2] # 입력 받은 시각의 승차 인원 값 추출
    if a > mx:
        mx = a
        mx_station = row[3] + '('+row[1]+')'
print(mx_station, mx)
```

#### 시간대별로 사람들이 가장 많이 타고 내리는 역은 어디일까

```python
import csv
# 3시간동안 가장 많이 들어오는 역 확인

import csv
f = open('data/subwaytime_2020.05.csv')
data = csv.reader(f)
next(data)
next(data)

mx = [0] * 24 # 시간대별 최대 승차인원 저장 리스트 초기화  
mx_station = [''] * 24  # 시간대별 최대 승차인원 역 이름 저장 리스트 초기화
for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[j*2+4] # 입력 받은 시각의 승차 인원 값 추출
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3] + '('+str(j+4)+'시)'
print(mx_station)
print(mx)
```

```python
# 그래프로 표현
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font', family = 'Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation = 90)
plt.show()
```

#### 시간대별로 하차 인원이 가장 많은 역을 찾는 코드

```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline

f = open('data/subwaytime_2020.05.csv')
data = csv.reader(f)
next(data)
next(data)

mx = [0] * 24  # 시간대별 최대 하차인원 저장 리스트 초기화
mx_station = [''] * 24  # 시간대별 최대 하차인원 역 이름 저장 리스트 초기화
for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[5 + j*2]  # 입력 받은 시각의 하차인원 값 추출
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3] + '('+str(j+5)+'시)'
plt.rc('font', family = 'Malgun Gothic')
plt.bar(range(24), mx, color = 'b')
plt.xticks(range(24), mx_station, rotation = 90)
plt.show()
```

#### 모든 지하철역에서 시간대별 승하차 인원을 모두 더하면?

```python
import csv

f = open('data/subwaytime_2020.05.csv')
data = csv.reader(f)
next(data)
next(data)
s_in = [0] * 24
s_out = [0] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4 + i * 2]
        s_out[i] += row[5 + i * 2]
```

#### 지하철 시간대별 승하차 인원 추이를 나타내는 코드

```python
import csv
import matplotlib.pyplot as plt

f = open('data/subwaytime_2020.05.csv')
data = csv.reader(f)
next(data)
next(data)
s_in = [0] * 24
s_out = [0] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4 + i * 2]
        s_out[i] += row[5 + i * 2]
        
plt.rc('font', family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```
